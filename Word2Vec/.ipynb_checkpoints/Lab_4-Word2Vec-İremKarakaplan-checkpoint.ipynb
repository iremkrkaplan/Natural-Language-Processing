```python
#Bu kod, Word2Vec modelini kullanarak bir metin kümesindeki (corpus) kelimeler arasındaki anlamsal ilişkileri,
#sayısal olarak öğrenme sürecini uygulamaktadır.
import gensim
from gensim.models import Word2Vec
import re
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
# is NLTK data (stopwords) downloaded
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')

# The file name def for the corpus
file_name = "teaching_learning_corpus.txt"
sentences = []

# Read the sentences from the text file
try:
    # Open our teaching_learning_corpus.txt file in read mode and read all lines from the file
    with open(file_name, 'r', encoding='utf-8') as f:
        raw_lines = f.readlines()
        
        # Clean and filter the lines: remove source tags and newlines
        # Iterate through raw lines to extract only the clean sentence text
        for line in raw_lines:
            
            # I encountered issues with extra spaces and newline characters when reading the file
            # I use .strip() here to solve the "empty space issue" and ensure a clean sentence
            # additionally, the regex removes the source tags (e.g., "")
            
            # Remove the source tag (like "") using a regular expression
            cleaned_line = re.sub(r'\\s*', '', line).strip()
            # Only include non-empty lines which are the actual sentences
            if cleaned_line:
                sentences.append(cleaned_line)

except FileNotFoundError:
    # file is missing
    print(f"ERROR: The file '{file_name}' was not found. Ensure it is in the same directory as the notebook.")

print(f"Corpus size (number of sentences): {len(sentences)}")
```

```python
# Tokenization

# Clean the text and perform tokenization (splitting into words)
tokenized_sentences = []
# Iterate through all sentences to clean and split them into words
for sentence in sentences:
    # Lowercase the text and remove all non-alphabetic characters
    clean_sentence = re.sub(r'[^a-zA-Z\s]', '', sentence.lower())
    # Split the cleaned sentence into individual word tokens
    tokens = clean_sentence.split()
    tokenized_sentences.append(tokens)

# the result of the tokenization process
print("\nFirst 3 tokenized sentences:")
print(tokenized_sentences[:3])

# ----Train the Word2Vec model ---

# Initialize and train the Word2Vec model using Gensim(word2vec algoritmalarını kullanabildiğimiz açık kaynak kütüp.)
model = Word2Vec(
    sentences=tokenized_sentences,
    vector_size=10,  # define the dimensionality of the word vectors
    window=5,        # set the context window size
    min_count=1,     # include all words present in the corpus
    sg=1,            # use the Skip-gram training algorithm (Skip-gram). Skip-gram'ı tercih ettim çünkü küçük bir korpusta, nadir kelimelerin (örneğin 'summative' veya 'pedagogical') bağlamlarını daha iyi öğrenme eğilimindedir.
    epochs=100       # Train the model over 100 iterations----Küçük korpuslarda doğru öğrenme için yüksek epochs kullanmamız gerekiyor. Modelin veriyi yeterince görebilmesi için
)

# Build and train our model using the tokenized sentences
model.build_vocab(tokenized_sentences)
model.train(tokenized_sentences, total_examples=model.corpus_count, epochs=model.epochs)

print(f"\nTotal words learned: {len(model.wv)}")

# Benzer Kelimeleri Bulma ve Vektör Çıktısı

# Keywords to test for semantic similarity
# !!! I noticed there is no teacher in the given txt so I deleted it from the keywords.
keywords = ["university", "student", "learning"]

# Print the top 10 most similar words for each keyword
# Iterate over all target words to find and display their nearest neighbors
for word in keywords:
    print(f"\n--- Top 10 Similar Words for: '{word}' ---")
    try:
        # Use the most_similar method to find semantically close words iki word arasındaki -cosine similarity-e bakarız
        similar_words = model.wv.most_similar(word, topn=10)
        # Print the similar words and their cosine similarity scores
        for similar_word, score in similar_words:
            print(f"- {similar_word} (Similarity: {score:.4f})")
    except KeyError:
        # Handle cases where the word is missing
        print(f"Error: The word '{word}' is not in the vocabulary.")

# Get the vector representation for the word "student"
#student kelimesi için vektör çıktısı istenerek, Word2Vec'in anlamsal bilgiyi nasıl sayısal bir diziye (vektöre) dönüştürdüğünü gösteriyoruz.
student_vector = model.wv['student']

# Print the first 10 numbers of the vector (Delivery: Vector for the word "student")
print(f"\nVector for 'student' (First 10 numbers):\n{student_vector[:10]}")
```

```python
# Prepare data for PCA visualization
# Extract all word labels and their corresponding high-dimensional vectors
vocab = list(model.wv.index_to_key)
X = model.wv[vocab]

# Apply PCA to reduce the dimensionality to 2 for plotting
# PCA helps project the 10D vectors into a 2D space while preserving most variance
pca = PCA(n_components=2)
result = pca.fit_transform(X)

# Create a scatter plot of the 2D vectors
plt.figure(figsize=(12, 10))
plt.scatter(result[:, 0], result[:, 1], alpha=0.7)

# Annotate each point with its corresponding word label
for i, word in enumerate(vocab):
    plt.annotate(word, xy=(result[i, 0], result[i, 1]), xytext=(5, 2), textcoords='offset points')

# Set the title and labels for clarity
plt.title("2D Visualization of Word Embeddings (PCA) - Teaching Corpus")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show() # 

# Choose a test word for semantic analysis: I choosed knowledge
test_word = "knowledge"
print(f"\n--- Analysis: 3 Nearest Neighbors for '{test_word}' ---")

try:
    # Find the top 3 nearest neighbors for the test word
    neighbors = model.wv.most_similar(test_word, topn=3)
    
    # Print the neighbors
    for word, score in neighbors:
        print(f"- {word} (Similarity: {score:.4f})")

    # Analyze if the 3 nearest neighbors make semantic sense and explain
    print("\nDo the 3 nearest neighbors make semantic sense? Explain:")
    print("> The neighbors for 'knowledge' (likely 'acquire', 'lectures', 'discussions' based on the corpus) reflect its immediate context in Sentence 2.")
    print("> Since 'Students acquire knowledge not only through lectures, but also through discussions...' is the only mention, the model learns strong co-occurrence with the surrounding words.")
    print("> This confirms that in a small, domain-specific corpus, semantic sense is *highly localized* and context-dependent, showing a limitation in capturing generalized meaning.")

except KeyError:
    print(f"Error: The test word '{test_word}' is not in the vocabulary.")

# Discuss the role of corpus size (Analysis: Discuss the role of corpus size)
print("\n--- Discussion: The Role of Corpus Size ---")
print("The **quality of Word2Vec embeddings** depends heavily on corpus size. A **large and diverse corpus** offers varied contexts, enabling the model to learn robust and generalizable semantic features.")
print("Conversely, this **small, specific corpus** produces **fragile vectors** that are primarily driven by local co-occurrence frequencies, rather than deep semantic relationships.")
print("The limited data means words with few mentions, like 'knowledge,' have shallow representations, hindering the model's ability to capture nuanced meaning outside the exact training sentences.")

#Analysis: Consider eliminating stop words and retry

# Define the set of English stop words
english_stop_words = set(stopwords.words('english'))

# Clean and tokenize the sentences, eliminating stop words
tokenized_sentences_no_stopwords = []
for tokens in tokenized_sentences:
    # Filter out any word that is in the predefined list of stop words
    filtered_tokens = [word for word in tokens if word not in english_stop_words]
    tokenized_sentences_no_stopwords.append(filtered_tokens)

print("\nTokenized sentences (No Stop Words):", tokenized_sentences_no_stopwords[:3])

# Train the Word2Vec model again using the filtered sentences
# A new model instance must be created for retraining
# I noticed overfitting occurs because similarity scores are so high between 0.96 - 0.99 and this is happening becaıse our corpus is so small
    
model_no_stopwords = Word2Vec(
    sentences=tokenized_sentences_no_stopwords,
    vector_size=10,
    window=5,
    min_count=1,
    sg=1,
    epochs=100
)
model_no_stopwords.build_vocab(tokenized_sentences_no_stopwords)
model_no_stopwords.train(tokenized_sentences_no_stopwords, total_examples=model_no_stopwords.corpus_count, epochs=model_no_stopwords.epochs)

# Find and print the top 10 most similar words for the keywords (after stop words removal)
print("\n--- Top 10 Similar Words (NO STOP WORDS) ---")
for word in keywords:
    print(f"\n--- Top 10 Similar Words for: '{word}' (No Stop Words) ---")
    try:
        similar_words = model_no_stopwords.wv.most_similar(word, topn=10)
        for similar_word, score in similar_words:
            print(f"- {similar_word} (Similarity: {score:.4f})")
    except KeyError:
        print(f"Error: The word '{word}' is not in the new vocabulary.")
```
